```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv('Artificats\\Time_Estimation_Dataset\\test.csv')
df_out = pd.read_csv('Artificats\\Time_Estimation_Dataset\\Sample_Submission.csv')
```

```python
df["Time_taken(min)"] = df_out["Time_taken(min)"]
```

```python
df
```

```python
def data(df):
    df.replace({"NaN": np.nan}, regex=True, inplace = True)
    df.drop(['Time_Order_picked','Delivery_person_ID','Restaurant_latitude','Restaurant_longitude','Type_of_order','Festival'],axis=1,inplace=True)
    # df['Time_taken(min)'] = df['Time_taken(min)'].str.replace('(min)','')
    df=df.rename(columns={'Time_taken(min)':'Time_taken'})
    df['Weatherconditions']=df['Weatherconditions'].str.replace('conditions','')
    df['Delivery_person_Age']=df['Delivery_person_Age'].astype(float)
    df['Delivery_person_Ratings']=df['Delivery_person_Ratings'].astype(float)
    df['Order_Date']=pd.to_datetime(df['Order_Date'],dayfirst=True)
    df['Time_Orderd'] = pd.to_datetime(df['Time_Orderd'], format='%H:%M:%S', errors='coerce').dt.time
    return df
```

```python
df = data(df)
```
